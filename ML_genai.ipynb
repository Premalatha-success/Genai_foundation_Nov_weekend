```python
df=df.drop("Loan_ID",axis=1)
```

```python
df.shape
```

```python
df_filled = df.copy()
df_filled['Gender'].fillna(df_filled['Gender'].mode()[0], inplace=True)
df_filled['Married'].fillna(df_filled['Married'].mode()[0], inplace=True)
df_filled['Self_Employed'].fillna(df_filled['Self_Employed'].mode()[0], inplace=True)
df_filled['LoanAmount'].fillna(df_filled['LoanAmount'].mean(), inplace=True)
df_filled['Loan_Amount_Term'].fillna(df_filled['Loan_Amount_Term'].mode()[0], inplace=True)
df_filled['Credit_History'].fillna(df_filled['Credit_History'].mode()[0], inplace=True)

df_filled.isnull().sum()
```

```python
df['Dependents'] = df['Dependents'].str.replace('+', '', regex=False)

df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
df['Married'] = df['Married'].map({'No': 0, 'Yes': 1})
df['Education'] = df['Education'].map({'Graduate': 1, 'Not Graduate': 0})
df['Self_Employed'] = df['Self_Employed'].map({'No': 0, 'Yes': 1})
df['Property_Area'] = df['Property_Area'].map({'Rural': 0, 'Semiurban': 1, 'Urban': 2})
df['Loan_Status'] = df['Loan_Status'].map({'N': 0, 'Y': 1})
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.dtypes
```

```python
df['Dependents'] = pd.to_numeric(df['Dependents'])
df['Dependents'].fillna(0, inplace=True)
df.dtypes
```

```python
Y=df[["Loan_Status"]]
X=df.drop("Loan_Status",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(X_train,Y_train)
```

```python
model_lr.score(X_train,Y_train)
```

```python
model_lr.score(X_test,Y_test)
```
