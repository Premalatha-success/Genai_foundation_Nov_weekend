```python
import pandas as pd
url = "https://raw.githubusercontent.com/Premalatha-success/Datasets/refs/heads/main/AutoInsurance.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Customer',"Effective To Date"], axis=1, errors='ignore')  # ignore errors if 'Customer' not present
# As per df.isnull().sum(), there are no missing values in the dataset, so no fillna is needed.
# Encode categorical variables
```

```python
categorical_cols = df.select_dtypes(include='object').columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
# Verify
df.head()
```

```python
X = df.drop('Customer Lifetime Value', axis=1)
y = df['Customer Lifetime Value']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3
)
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
lr_model.score(X_train, y_train)
```

```python
lr_model.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
model_dt=DecisionTreeRegressor()
model_dt.fit(X_train,y_train)
```

```python
model_dt.score(X_train,y_train)
```

```python
model_dt.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rf=RandomForestRegressor()
model_rf.fit(X_train,y_train)
model_rf.score(X_train,y_train)
```

```python
model_rf.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingRegressor
model_bg=BaggingRegressor()
model_bg.fit(X_train,y_train)
model_bg.score(X_train,y_train)
```

```python
model_bg.score(X_test,y_test)
```
